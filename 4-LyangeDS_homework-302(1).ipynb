#### Практика

```python
import pandas
import numpy
import pylab
%matplotlib inline
import matplotlib.pyplot as plt

from sklearn.preprocessing import scale
```

#### Реализация алгоритма обучения персептрона с фиксированной скоростью обучения и порядком обхода

```python
trainFile = pandas.read_csv('train.csv')

extractedClassTrain = trainFile['class']
extractedClassTrainArray = numpy.array(extractedClassTrain)
trainFileWithoutClass = trainFile.drop('class', 1)
trainFileWithoutClassArray = numpy.array(trainFileWithoutClass)
normalizedTrainFileWithoutClass = scale(trainFileWithoutClass.values)
normalizedTrainFileWithoutClass = pandas.DataFrame(normalizedTrainFileWithoutClass, columns=['preg', 'plas', 'pres', 'skin', 'insu', 'mass', 'pedi', 'age'])
normalizedTrainFile = normalizedTrainFileWithoutClass
normalizedTrainFile['class'] = extractedClassTrain

#print normalizedTrainFile

testFile = pandas.read_csv('test.csv')

extractedClassTest = testFile['class']
extractedClassTestArray = numpy.array(extractedClassTest)
testFileWithoutClass = testFile.drop('class', 1)
testFileWithoutClassArray = numpy.array(testFileWithoutClass)
normalizedTestFileWithoutClass = scale(testFileWithoutClass.values)
normalizedTestFileWithoutClass = pandas.DataFrame(normalizedTestFileWithoutClass, columns=['preg', 'plas', 'pres', 'skin', 'insu', 'mass', 'pedi', 'age'])
normalizedTestFileWithoutClass['class'] = extractedClassTest
normalizedTestFile = normalizedTestFileWithoutClass

#print normalizedTestFile

constantSpeedOfLearningArray = [0.0001, 0.001, 0.01, 0.1]
# лучший вариант всегда получался для 0.0001

# взял рандомно сгенерированный вектор параметров, а не 0-ой
vectorOfParametersOriginal = numpy.array((numpy.random.rand(8,1)*2-1)/len(normalizedTrainFile))
#vectorOfParametersOriginal = [0, 0, 0, 0, 0, 0, 0, 0]
# выбран ненулевой вектор, так как для него минимальные проценты ошибок достигались быстрее, но в моих тестах разница была совсем мала,
# поэтому не могу привести убедительного примера. Для ненулевого векторадля 0.0001 скорости идеал в одном из тестов 
# достигался на 16 итерации, а для нулевого - на 40-ой. Для
notSoRandomConstantOriginal = (numpy.random.rand()*2-1)/len(normalizedTrainFile)

amountOfMistakesForAllCyclesForDifferentSpeedOfLearning = list()

# Далее пойдет объяснение, почему было выбрано конкретное значение итераций (500), а не описанная в лекции
# остановка работы программы в случае превышения процента ошибок на данном ходу (итерации) по сравнению с предыдущим.
# В начале я действительно реализовал такой способ работы, но потом оказалось, что существует множество локальных минимумов,
# на которых работы программы останавливается, не достигнув оптимального значения. 
# Поэтому было принято решение взять заведомо большее число циклов, чем нужно, для нахождения их оптимального числа 
# (определил в ходе запусков для разного числа итераций - 10, 50, 100 и тд. до 1000)

numberOfCycles = 500 #итерации
for constantSpeedOfLearning in constantSpeedOfLearningArray:
    vectorOfParameters = list(vectorOfParametersOriginal)
    notSoRandomConstant = notSoRandomConstantOriginal
    amountOfMistakesForAllCycles  = list()
    for cycle in xrange(numberOfCycles):
        for trainObjectId in xrange(len(trainFileWithoutClassArray)):
            while True:
                multiplication = numpy.dot(numpy.array(trainFileWithoutClassArray[trainObjectId]), vectorOfParameters)
                result = numpy.sum(multiplication) + notSoRandomConstant
                if ((result > 0 and extractedClassTrainArray[trainObjectId] == 1) or (result < 0 and extractedClassTrainArray[trainObjectId] == -1)):
                    #удовлетворяет условию, прерываем
                    break
                else:
                    #изменяем параметры (с учетом нужного знака) 
                    for parameter in xrange(len(vectorOfParameters)):
                        vectorOfParameters[parameter] += constantSpeedOfLearning*trainFileWithoutClassArray[trainObjectId][parameter]*extractedClassTrainArray[trainObjectId]
                        notSoRandomConstant -= constantSpeedOfLearning*extractedClassTrainArray[trainObjectId]

        amountOfMistakesForCycle = 0
        for testObjectId in xrange(len(testFileWithoutClassArray)):
            multiplication = numpy.dot(numpy.array(testFileWithoutClassArray[testObjectId]), vectorOfParameters)
            result = numpy.sum(multiplication)
            if ((result > 0 and extractedClassTestArray[testObjectId] == -1) or (result < 0 and extractedClassTestArray[testObjectId] == 1)):
                amountOfMistakesForCycle += 1

        amountOfMistakesForAllCycles.append(amountOfMistakesForCycle*100/len(testFileWithoutClassArray))
    
    amountOfMistakesForAllCyclesForDifferentSpeedOfLearning.append(amountOfMistakesForAllCycles) 

for amountOfMistakesForAllCyclesForOneSpeedOfLearning in amountOfMistakesForAllCyclesForDifferentSpeedOfLearning:
    plt.plot(xrange(len(amountOfMistakesForAllCyclesForOneSpeedOfLearning)), amountOfMistakesForAllCyclesForOneSpeedOfLearning)
    plt.ylabel('Mistakes, %')
    plt.xlabel('Iterations')
    plt.show()
```

#### Реализация алгоритма обучения персептрона с переменной скоростью обучения 

```python
vectorOfParameters = numpy.array((numpy.random.rand(8,1)*2-1)/len(normalizedTrainFile))

notSoRandomConstant = (numpy.random.rand()*2-1)/len(normalizedTrainFile)


numberOfCycles = 400 #итерации

randomTrainObjectIdArray = numpy.arange(len(trainFileWithoutClassArray))
numpy.random.shuffle(randomTrainObjectIdArray)

amountOfMistakesForAllCycles = list()

for cycle in xrange(numberOfCycles):
    speedOfLearning = float(1)/(cycle + 1)
    for randomTrainObjectId in randomTrainObjectIdArray:
        while True:
            multiplication = numpy.dot(numpy.array(trainFileWithoutClassArray[randomTrainObjectId]), vectorOfParameters)
            result = numpy.sum(multiplication) + notSoRandomConstant
            if ((result > 0 and extractedClassTrainArray[randomTrainObjectId] == 1) or (result < 0 and extractedClassTrainArray[randomTrainObjectId] == -1)):
                #удовлетворяет условию, прерываем
                break
            else:
                #изменяем параметры (с учетом нужного знака) 
                for parameter in xrange(len(vectorOfParameters)):
                    vectorOfParameters[parameter] += speedOfLearning*trainFileWithoutClassArray[randomTrainObjectId][parameter]*extractedClassTrainArray[randomTrainObjectId]
                    notSoRandomConstant -= speedOfLearning*extractedClassTrainArray[randomTrainObjectId]

    amountOfMistakesForCycle = 0
    for testObjectId in xrange(len(testFileWithoutClassArray)):
        multiplication = numpy.dot(numpy.array(testFileWithoutClassArray[testObjectId]), vectorOfParameters)
        result = numpy.sum(multiplication)
        if ((result > 0 and extractedClassTestArray[testObjectId] == -1) or (result < 0 and extractedClassTestArray[testObjectId] == 1)):
            amountOfMistakesForCycle += 1

    amountOfMistakesForAllCycles.append(amountOfMistakesForCycle*100/len(testFileWithoutClassArray))


plt.plot(xrange(len(amountOfMistakesForAllCycles)), amountOfMistakesForAllCycles)
plt.ylabel('Mistakes, %')
plt.xlabel('Iterations')
plt.show()

# Результаты
# К сожалению, минимальное число ошибок составило 33%
# Уменьшающаяся с числом итераций скорость обучения привела к сокращению роста разрыва колебаний при увеличеии числа итераций.

# Опять же, для ненулевого вектора результаты оказались лучше
```

```python
u1 = [3, 3]
u2 = [1, 1]
cov_matrix = [[1,0],[0,1]]
size = 100
first_selection = numpy.random.multivariate_normal(u1, cov_matrix, size) 
second_selection = numpy.random.multivariate_normal(u2, cov_matrix, size)
```

```python
arrayOfOnes = numpy.ones((100, 1))
arrayOfMinusOnes = numpy.ones((100, 1))

for ones in arrayOfMinusOnes:
    ones[0] = -1

first_selection_copy = numpy.hstack((first_selection, arrayOfOnes))
second_selection_copy = numpy.hstack((second_selection, arrayOfMinusOnes))

allData = numpy.vstack((first_selection_copy, second_selection_copy))
allDataWithoutClasses = numpy.vstack((first_selection, second_selection))

numpy.random.shuffle(allData)
```

```python
import sklearn
from sklearn import neighbors, datasets, tree
X = numpy.concatenate((first_selection, second_selection), axis=0)
Y = numpy.concatenate(([0] * len(first_selection), [1] * len(second_selection)), axis = 0) # task 2

h = .02 
x_min, x_max = X[:,0].min() - .5, X[:,0].max() + .5
y_min, y_max = X[:,1].min() - .5, X[:,1].max() + .5
xx, yy = numpy.meshgrid(numpy.arange(x_min, x_max, h), numpy.arange(y_min, y_max, h))

tree = sklearn.tree.DecisionTreeClassifier()
tree.fit(X, Y)
prediction2 = tree.predict(numpy.c_[xx.ravel(), yy.ravel()])

def print_result(prediction):
    prediction = prediction.reshape(xx.shape)
    pylab.figure(1, figsize=(4, 3))
    pylab.set_cmap(pylab.cm.summer_r)
    pylab.pcolormesh(xx, yy, prediction)

    pylab.scatter(X[:,0], X[:,1], c=Y)

    pylab.xlim(xx.min(), xx.max())
    pylab.ylim(yy.min(), yy.max())
    pylab.xticks(())
    pylab.yticks(())

    pylab.show()
    
print_result(prediction2)
```

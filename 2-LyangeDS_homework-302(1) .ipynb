#### Questions 1-3

```python
def generate_points_of_this_dimension(number, dimension):
    points_array = []
    for i in xrange(number):
        generation_iteration = numpy.random.uniform(-1, 1, dimension)
        points_array.append(generation_iteration)
    return points_array

def calculate_cos(points_array_1, points_array_2):
    p1_length = numpy.linalg.norm(points_array_1, ord=2)
    p2_length = numpy.linalg.norm(points_array_2, ord=2)
    length = len(points_array_1)
    points_result = 0
    for i in xrange(length):
        points_result += points_array_1[i] * points_array_2[i]
    return points_result / (p1_length * p2_length)

average_min_cos_values_array = []
average_max_cos_values_array = []

for dimension in xrange(1, 201):
    points_of_this_dimension_array = generate_points_of_this_dimension(100, dimension)
    min_cos_for_this_dimension_array = []
    max_cos_for_this_dimension_array = []
    for i in xrange(len(points_of_this_dimension_array)):
        min_cos = 1.0
        max_cos = -1.0
        for j in xrange(len(points_of_this_dimension_array)):
            if i != j:
                cos = calculate_cos(points_of_this_dimension_array[i], points_of_this_dimension_array[j])
                if cos < min_cos:
                    min_cos = cos
                if cos > max_cos:
                    max_cos = cos
        min_cos_for_this_dimension_array.append(min_cos)
        max_cos_for_this_dimension_array.append(max_cos)
    average_min_cos_values_array.append(numpy.average(min_cos_for_this_dimension_array))
    average_max_cos_values_array.append(numpy.average(max_cos_for_this_dimension_array))

fig, (ax0, ax1) = matplotlib.pyplot.subplots(nrows=2)
ax0.plot(xrange(1, 201), average_max_cos_values_array)
ax0.set_title('average maximum cosinus values')
ax1.plot(xrange(1, 201), average_min_cos_values_array)
ax1.set_title('average minimum cosinus values')
matplotlib.pyplot.subplots_adjust(hspace=0.5) 
matplotlib.pyplot.show()
```

#### Questions 4-7

```python
def set_new_values(new, old_min, old_max, is_set_already): 
    new_min = old_min
    new_max = old_max
    if is_set_already:
        if old_min > new:
            new_min = new
        elif old_max < new:
            new_max = new
    else:
        is_set_already = True
        new_max = new
        new_min = new
    return new_min, new_max, is_set_already

def calculate_norm(type):
    l_array = []
    for dimension in xrange(1, 1001):
        points_of_this_dimension_array = generate_points_of_this_dimension(1000, dimension)

        l_max = 0
        l_min = 0
        l_is_set_already = False

        for point in points_of_this_dimension_array:
            l_new = numpy.linalg.norm(point, ord=type)
            l_min, l_max, l_is_set_already = set_new_values(l_new, l_min, l_max, l_is_set_already)

        l_difference = (l_max - l_min) / l_max
        l_array.append(l_difference)
    return l_array


l2_array = calculate_norm(1)
l1_array = calculate_norm(2)
l8_array = calculate_norm(numpy.inf)
```

```python
pylab.plot(l2_array)
pylab.xlabel('Dimension')
pylab.ylabel('')
pylab.title('L2')
```

```python
pylab.plot(l1_array)
pylab.xlabel('Dimension')
pylab.ylabel('')
pylab.title('L1')
```

```python
pylab.plot(l8_array)
pylab.xlabel('Dimension')
pylab.ylabel('')
pylab.title('Linfinity')
```

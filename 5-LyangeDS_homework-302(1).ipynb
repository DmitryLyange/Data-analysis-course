```python
import pandas
import numpy
import pylab
%matplotlib inline
import matplotlib.pyplot as plt

import sklearn.svm
from numpy import logspace, meshgrid
```

###### learn_stat

```python
# Следующий блок кода я отправлял в сдаче 6 лабороторной, за что мне и поставили 0,6*3/5 = 0.4. Пишу, дабы Вам не пришлось перепроверять,
# почему там стоит такой балл.
# Заранее извиняюсь за возможное неудобство чтения кода, так как в условиях заданий переменные и функции пишутся через нижнее подчеркивание, 
# а я постоянно переключаюсь на camelCase

def learn_stat(train_data, train_labels, test_data, test_labels, kernel, C):
    svc = sklearn.svm.SVC(C = C, kernel = kernel)
    svc.fit(train_data, train_labels)
    train_score = svc.score(train_data, train_labels)
    test_score = svc.score(test_data, test_labels)
    support = len(svc.support_)
    return train_score, test_score, support

# Блок кода из уже сданного завершен
```

###### plot_stat

```python
def plot_stat(train_data, train_labels, test_data, test_labels, kernel, minC, maxC, steps):

    # numpy.logspace(start, stop, num=50, endpoint=True, base=10.0, dtype=None)[source]
    logMinC = numpy.log(minC)
    logMaxC = numpy.log(maxC)
    eLog = numpy.math.e # основание
    CArray = logspace(logMinC, logMaxC, num = steps, base = eLog, dtype = None)

    train_scoreArray = list()
    test_scoreArray = list()
    supportArray = list()

    for CItem in CArray:
        train_score, test_score, support = learn_stat(train_data, train_labels, test_data, test_labels, kernel, CItem)
        train_scoreArray.append(train_score)
        test_scoreArray.append(test_score)
        supportArray.append(support)
    
    # переводим в логарифмическую шкалу от С
    for CId in xrange(len(CArray)):
        CArray[CId] = numpy.log(CArray[CId])


    plt.plot(CArray, train_scoreArray)
    plt.xlabel('log(C)')
    plt.ylabel('Accuracy')
    if (kernel == "linear"):
        # для лучшей видимости прямых линий, совпадающих иначе с границами
        plt.ylim(0.814, 0.861)
    plt.show()
    
    plt.plot(CArray, test_scoreArray)
    plt.xlabel('log(C)')
    plt.ylabel('Accuracy')
    if (kernel == "linear"):
        plt.ylim(0.814, 0.861)
    plt.show()
    
    plt.plot(CArray, supportArray)
    plt.xlabel('log(C)')
    plt.ylabel('Support')
    plt.show()
```

###### plot_support

```python
def plot_support(train_data, train_labels, test_data, test_labels, kernel, C):
    svc = sklearn.svm.SVC(kernel = kernel, C = C)
    svc.fit(train_data, train_labels)
    
    supportVectors = svc.support_vectors_
    support = svc.support_
    
    x_min = test_data[:, 0].min()
    x_max = test_data[:, 0].max() + 0.5
    y_min = test_data[:, 1].min()
    y_max = test_data[:, 1].max() + 0.5
    # 0.5 - step 
    xVector = numpy.arange(x_min, x_max, 0.5)
    yVector = numpy.arange(y_min, y_max, 0.5)
    
    XMatrice, YMatrice = meshgrid(xVector, yVector)
 
    XReshaped = XMatrice.reshape(-1)
    YReshaped = YMatrice.reshape(-1)
    
    # XYCombined = numpy.hstack((XReshaped, YReshaped)) - не работает TODO
    XYCombined = numpy.c_[XReshaped, YReshaped]
    
    # svc.decision_function(X) or svc.predict(X) TODO
    
    #Z = svc.predict(XYCombined)
    Z = svc.decision_function(XYCombined)
    Z = Z.reshape(XMatrice.shape)
    
    plt.figure()
    plt.plot(xticks=[], yticks=[])

    # plt.scatter(x, y, s=area, c=colors, alpha=0.5)
    
    plt.scatter(test_data[:, 0], test_data[:, 1], c = test_labels)
    plt.scatter(supportVectors[:, 0], supportVectors[:, 1], c=train_labels[(numpy.array(support))], cmap=plt.cm.cool)
    
    # example from countourf_demo
    # CS = plt.contourf(X, Y, Z, 10, # [-1, -0.1, 0, 0.1],
    #                    #alpha=0.5,
    #                    cmap=plt.cm.bone,
    #                    origin=origin)
    

    if (kernel == 'linear'):
        plt.contour(XMatrice, YMatrice, Z, levels = [-1.0, 0.0, 1.0])
    else:
        plt.contourf(XMatrice, YMatrice, Z, 15,
                    cmap = plt.cm.spring,
                    alpha=0.4,
                    zorder = 0)

    plt.show()
```

###### 3 пункт задания

```python
def bestC(train_data, train_labels, test_data, test_labels, kernel, minC, maxC, steps):

    logMinC = numpy.log(minC)
    logMaxC = numpy.log(maxC)
    eLog = numpy.math.e # основание
    CArray = logspace(logMinC, logMaxC, num = steps, base = eLog, dtype = None)

    bestScore = 0
    bestC = 0

    for CItem in CArray:
        train_score, test_score, support = learn_stat(train_data, train_labels, test_data, test_labels, kernel, CItem)

        if test_score > bestScore:
            bestC = CItem
            bestScore = test_score
            
    return bestC
        
bestCLinear = bestC(trainWithoutLabel, trainLabel, testWithoutLabel, testLabel, 'linear', 1e-2, 200, 30 )    
bestCRbf = bestC(trainWithoutLabel, trainLabel, testWithoutLabel, testLabel, 'rbf', 1e-2, 200, 30 )

print bestCLinear
print bestCRbf

bestCLinear = bestC(trainWithoutLabel, trainLabel, testWithoutLabel, testLabel, 'linear', 1e-2, 10000, 30 )    
bestCRbf = bestC(trainWithoutLabel, trainLabel, testWithoutLabel, testLabel, 'rbf', 1e-2, 10000, 30 )

print bestCLinear
print bestCRbf

print "При ограничении на maxC, наложенном на 2 задачу, лучшее bestCRbf для входящих данных, равное 221.221629107, не достигается."
print "Но если ограничения сверху снять, то лучшее bestCRbf достигается."
```

```python
print "Неадекватно маленькое С"
plot_support(trainWithoutLabelArray, trainLabelArray, testWithoutLabelArray, testLabelArray, 'linear', 0.001)
plot_support(trainWithoutLabelArray, trainLabelArray, testWithoutLabelArray, testLabelArray, 'rbf', 0.001)

print "Неадекватно большое С"
plot_support(trainWithoutLabelArray, trainLabelArray, testWithoutLabelArray, testLabelArray, 'linear', 100)
plot_support(trainWithoutLabelArray, trainLabelArray, testWithoutLabelArray, testLabelArray, 'rbf', 100)

print "Наилучшее С"
plot_support(trainWithoutLabelArray, trainLabelArray, testWithoutLabelArray, testLabelArray, 'linear', bestCLinear)
plot_support(trainWithoutLabelArray, trainLabelArray, testWithoutLabelArray, testLabelArray, 'rbf', bestCRbf)
```

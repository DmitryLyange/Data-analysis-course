```python
grass = imread('grass.jpg')

# каркас взят с plot_lena_segmentation.py

grass = numpy.dot(grass[..., :3], [0.299, 0.587, 0.144])

graph = image.img_to_graph(grass)

beta = 5
eps = 1e-6
graph.data = numpy.exp(-beta * graph.data / grass.std()) + eps

clustersAmountArray = [2, 5, 10, 15, 20, 25]

for clustersAmount in clustersAmountArray:
    for assign_labels in ('kmeans', 'discretize'):

        labels = spectral_clustering(graph, n_clusters = clustersAmount,
                                     assign_labels=assign_labels,
                                     eigen_solver='amg')
        labels = labels.reshape(grass.shape[0], grass.shape[1])

        plt.figure(figsize=(5, 5))
        plt.imshow(grass,   cmap=plt.cm.gray)
        for l in range(clustersAmount):
            plt.contour(labels == l, contours=1,
                        colors=[plt.cm.spectral(l / float(clustersAmount)), ])
        plt.xticks(())
        plt.yticks(())
        plt.title('Spectral: {0}, K = {1}'.format(assign_labels, clustersAmount))
        plt.show()
```

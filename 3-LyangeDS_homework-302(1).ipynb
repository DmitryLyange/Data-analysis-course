#### Визуализация разделяющих поверхностей

```python
import pandas
import numpy
import pylab
%matplotlib inline
```

##### Задание 1

```python
u1 = [3, 3]
u2 = [1, 1]
cov_matrix = [[1,0],[0,1]]
size = 150

first_selection = numpy.random.multivariate_normal(u1, cov_matrix, size) 
second_selection = numpy.random.multivariate_normal(u2, cov_matrix, size)

pylab.scatter(first_selection[:,0], first_selection[:,1], color = 'r')
pylab.scatter(second_selection[:,0], second_selection[:,1], color = 'b')
```

##### Задание 2

```python
import sklearn
from sklearn import neighbors, datasets, tree
X = numpy.concatenate((first_selection, second_selection), axis=0)
Y = numpy.concatenate(([0] * len(first_selection), [1] * len(second_selection)), axis = 0) # task 2

h = .02 
x_min, x_max = X[:,0].min() - .5, X[:,0].max() + .5
y_min, y_max = X[:,1].min() - .5, X[:,1].max() + .5
xx, yy = numpy.meshgrid(numpy.arange(x_min, x_max, h), numpy.arange(y_min, y_max, h))
    
  
knn = sklearn.neighbors.KNeighborsClassifier()
knn.fit(X, Y)
prediction1 = knn.predict(numpy.c_[xx.ravel(), yy.ravel()])

tree = sklearn.tree.DecisionTreeClassifier()
tree.fit(X, Y)
prediction2 = tree.predict(numpy.c_[xx.ravel(), yy.ravel()])

def print_result(prediction):
    prediction = prediction.reshape(xx.shape)
    pylab.figure(1, figsize=(4, 3))
    pylab.set_cmap(pylab.cm.summer_r)
    pylab.pcolormesh(xx, yy, prediction)

    pylab.scatter(X[:,0], X[:,1], c=Y)

    pylab.xlim(xx.min(), xx.max())
    pylab.ylim(yy.min(), yy.max())
    pylab.xticks(())
    pylab.yticks(())

    pylab.show()
    
print_result(prediction1)
print_result(prediction2)
```

#### Сравнение качества kNN и рещающего дерева

##### Задание 1-5

```python
import matplotlib.pyplot
import sklearn
from sklearn import neighbors, datasets, tree
from mpl_toolkits.mplot3d import Axes3D

data = pandas.read_csv("adult.data", header = None, delimiter=',')
features = pandas.read_csv("adult.names", header = None, delimiter='\:')

continious_features = numpy.array(numpy.where(features == ' continuous.')) - 1
continuous_features = numpy.append(continious_features[0], 14)

data[14] = data[14].map({" <=50K": 0, ' >50K': 1})
continuous_data = data.as_matrix(continuous_features).astype(int)
train, test = sklearn.cross_validation.train_test_split(continuous_data, test_size = 0.3)

param = len(train[0]) - 1
train_1 = numpy.delete(train, param , axis = 1)
train_2 = numpy.delete(train, range(0, param), axis = 1)
test_x = numpy.delete(test, param, axis = 1)
test_y = numpy.delete(test, range(0, param), axis = 1)

tree = sklearn.tree.DecisionTreeClassifier()
depths = []
leaves = []
scores = []
for max_depth in xrange(1, 50) :
    for min_samples_leaf in xrange(1, 30) :
        tree.max_depth = max_depth
        tree.min_samples_leaf = min_samples_leaf
        tree.fit(train_1, train_2)
        depths.append(max_depth)
        leaves.append(min_samples_leaf)
        scores.append(tree.score(test_x, test_y))  
score = max(scores)
score_index = scores.index(max(scores))
min_sapmples_leaf = leaves[score_index]
max_depth = depths[score_index]


figure = matplotlib.pyplot.figure()
ax = figure.add_subplot(111, projection='3d')
ax.scatter(depths, leaves, scores, c='y', marker='o')
ax.set_xlabel('max_depth')
ax.set_ylabel('min_sapmples_leaf')
ax.set_zlabel('scores')
 
print("Score = " + str(score))
print("Max_depth = " + str(max_depth))
print("Min_samples_leaf = " + str(min_sapmples_leaf))
    
matplotlib.pyplot.show()
```
